# 12. Evaluation Metrics

Since the RainCast project is a classification problem, classification-based performance metrics are used to evaluate the models.

## Metrics Used

### 1. Accuracy
Accuracy measures the overall percentage of correct predictions made by the model.

**Formula:**
Accuracy = (Correct Predictions / Total Predictions)

### 2. Precision
Precision measures how many predicted positive cases were actually positive.

In this project:
- Out of all cases predicted as rain, how many were actually rainy days?

### 3. Recall
Recall measures how many actual positive cases were correctly identified.

In this project:
- Out of all actual rainy days, how many were correctly predicted by the model?

### 4. F1-Score
F1-score is the harmonic mean of Precision and Recall.

It is useful when:
- Class distribution may not be perfectly balanced
- We want a balanced performance measure

### 5. Classification Report
The classification report gives:
- Precision
- Recall
- F1-score
- Support

### 6. Confusion Matrix
The confusion matrix shows:
- True Positives
- True Negatives
- False Positives
- False Negatives

It helps us understand where the model is making mistakes.

## Why These Metrics Are Important
A rain forecasting system should not only be accurate, but also reliable in identifying rainy days correctly. Therefore, using multiple evaluation metrics provides a more complete understanding of model performance.

## Step 2.7: Standardize Categorical Values

Sometimes categorical values may contain inconsistent text such as:
- `yes`
- `YES`
- `No `
- ` no`

To avoid problems during feature encoding and model training, we standardize the text values.

```python
# Standardize text values
df['RainToday'] = df['RainToday'].astype(str).str.strip().str.capitalize()
df['RainTomorrow'] = df['RainTomorrow'].astype(str).str.strip().str.capitalize()

```

## Step 2.8: Handle Missing Values

In this step:
- Numerical columns are filled using **median**
- Categorical columns are filled using **mode**

This is a safe and commonly used data cleaning strategy for machine learning projects.

```python
# Fill missing values

# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled successfully.")
```

## Step 2.9: Final Missing Value Check

After filling missing values, we verify that the dataset is now clean and does not contain null values.

```python
# Final missing value check
df.isnull().sum()
```
